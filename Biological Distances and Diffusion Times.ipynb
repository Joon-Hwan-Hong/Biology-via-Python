```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def time(log_space, diffusion_constant):
    # time = length^2 / Diffusion constant. 
    return (log_space**2)/diffusion_constant

# so the documents say seq start is base ** start and ends with base ** stop
c = np.logspace(0, 3, num=1000) 

# Diffusion Constants. 
# 0. Protein in water diffusion constant of 100 1. Potassium in water 2. FP in E.coli Cytoplasm 3. DNA in yeast nucleus
DC0 = 100
DC1 = 2000 
DC2 = 7
DC3 = 5*10**(-4)

time0 = time(c, DC0)
time1 = time(c, DC1)
time2 = time(c, DC2)
time3 = time(c, DC3)
```

```python
plt.plot(c, time0, label='Protein in water')
plt.plot(c, time1, label='Potassium in water')
plt.plot(c, time2, label='GFP in E.coli cytoplasm')
plt.plot(c, time3, label='DNA in yeast nucleus')

# visual corrections
plt.xlim((1, 1e3))
plt.ylim(10**(-2), 10**4)
plt.yscale('log')
plt.xscale('log')

# labeling
plt.xlabel('length (μm)')
plt.ylabel('time (s)')
plt.legend()

plt
```

```python
# Stokes–Einstein equation
def Stokes_Einstein(temp, radius, viscosity):
    """
    :param Temp: Temperature in K
    :param radius: Radius of the spherically assumed object
    :param viscosity: The dynamic viscosity
    :return: Diffusion constant
    """
    kb = 1.380649*(10**(-23))
    return (kb*temp)/(6*radius*viscosity*math.pi)

def viscosity_radius_const(diffusion_const, temp):
    kb = 1.380649*(10**(-23))
    return kb*temp / (diffusion_const * 6 * math.pi)

print( Stokes_Einstein(298.15, 2.27*(10**(-10)), 0.00089) )
print(viscosity_radius_const(2000,298.15))
```
