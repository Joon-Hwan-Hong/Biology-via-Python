Shows the probabilities of simplified states available to dimoglobin in its interactions with oxygen binding partners.
Where there are 3 states, no occupancy, one occupancy in either active site, and both sites full.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# TODO: These are with set parameters of d(epsilon) and J, and C0. Make it modular later
```

```python
c = np.logspace(-2, 2, 1e3) # Array of 1000 logarithmically (in base 10) spaced numbers from 10^-2 to 10^2 (inclusively)

P0 = 1/(1 + 2 * np.exp(5) * c/760 + np.exp(12.5) * (c/760)**2)
P1 = 2 * np.exp(5) * c/760 * P0
P2 = np.exp(12.5) * (c/760)**2 * P0

plt.plot(c, P0, label = "P0") # Plots P0 as a function of c
plt.plot(c, P1, label = "P1") # Plots P1 as a function of c
plt.plot(c, P2, label = "P2") # Plots P2 as a function of c

plt.xscale("log") # Sets log scale
plt.xlim((1e-2, 1e2)) # Sets scale limits
plt.xlabel("Oxygen partial pressure")
plt.ylabel("Probability")
plt.legend()

plt.show()

```
