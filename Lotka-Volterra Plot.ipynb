```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N1_domain = np.linspace(0,50)
N2_domain = np.linspace(0,50)

# carrying capacity
K1 = 27.3
K2 = 35

# competition coefficient
alpha = 3.5
beta = 2.7

# linear competitive exclusion functions
def N1func(K1, alpha, N2):
    return (K1 - alpha*N2)

def N2func(K2, beta, N1):
    return (K2 - beta*N1)

N1F = N1func(K1, alpha, N1_domain)
N2F = N2func(K2, beta, N2_domain)
```

```python
plt.plot(N1F, N1_domain, label="Species 1")
plt.plot(N2_domain,N2F, label="Species 2")

plt.xlim(0,40)
plt.ylim(0,40)

plt.xlabel("N1")
plt.ylabel("N2")
plt.legend()

plt
```

```python
# CASE where an environmental factor decreases the competitive effect of species 1 on 2 to 0.8.

# altered competition coefficient
beta = 0.8

# linear competitive exclusion functions
N1F = N1func(K1, alpha, N1_domain)
N2F = N2func(K2, beta, N2_domain)

plt.plot(N1F, N1_domain, label="Species 1")
plt.plot(N2_domain,N2F, label="Species 2")

plt.xlim(0,50)
plt.ylim(0,40)
plt.xlabel("N1")
plt.ylabel("N2")
plt.legend()

plt
```
